```python
def combinations_check(record_choices):
    
    X_choices = [item for item in record_choices if record_choices.index(item)%2 == 0]
    
    X_num = 0
    
    X_record = []
    
    O_choices = [item for item in record_choices if record_choices.index(item)%2 == 1]

    O_num = 0
    
    O_record = []
    
    
    #chech X-rows
    for i in [0,1,2]:
        
        for j in [0,1,2]:
            
            if (i,j) in X_choices:
                
                X_num += 1
                
            else:
                
                X_num += 0
                
                
            
        X_record.append(X_num)
        
        X_num = 0
        
        
    #check O-rows    
    for i in [0,1,2]:
        
        for j in [0,1,2]:
            
            if (i,j) in O_choices:
                
                O_num += 1
                
            else:
                
                O_num += 0
                
                
            
        O_record.append(O_num)
        
        O_num = 0
        
        
        
        
    #check X-columns    
    for j in [0,1,2]:
        
        for i in [0,1,2]:
            
            if (i,j) in X_choices:
                
                X_num += 1
                
            else:
                
                X_num += 0
                
                
            
        X_record.append(X_num)
        
        X_num = 0
        
        
        
    #check O-columns    
    for j in [0,1,2]:
        
        for i in [0,1,2]:
            
            if (i,j) in O_choices:
                
                O_num += 1
                
            else:
                
                O_num += 0
                
                
            
        O_record.append(O_num)
        
        O_num = 0
    
    
    #check X-diagonal
    for i in [0,1,2]:
        
        if (i,i) in X_choices:
                
            X_num += 1
                
        else:
                
            X_num += 0
                
                
            
    X_record.append(X_num)
        
    X_num = 0
    
    
    #check O-diagonal
    for i in [0,1,2]:
        
        if (i,i) in O_choices:
                
            O_num += 1
                
        else:
                
            O_num += 0
                
                
            
    O_record.append(O_num)
        
    O_num = 0
        
    
    
    #check X-antidiagonal
    for i in [0,1,2]:
        
        if (i,2-i) in X_choices:
                
            X_num += 1
                
        else:
                
            X_num += 0
                
                
            
    X_record.append(X_num)
        
    X_num = 0
    
    
    #check O-antidiagonal
    for i in [0,1,2]:
        
        if (i,2-i) in O_choices:
                
            O_num += 1
                
        else:
                
            O_num += 0
                
                
            
    O_record.append(O_num)
        
    O_num = 0
    
    
    
    
    return X_record, O_record    
```
